```python
import multiprocessing as mp
import time

# Function to classify traffic
def process_lane(cars):
    if cars > 20:
        status = "High Traffic"
    elif cars > 10:
        status = "Medium Traffic"
    else:
        status = "Low Traffic"
    return (cars, status)

# Sequential processing
def sequential(lanes):
    result = []
    for lane in lanes:
        result.append(process_lane(lane))
    return result

# Parallel processing using multiprocessing
def parallel(lanes):
    pool = mp.Pool(mp.cpu_count())
    result = pool.map(process_lane, lanes)
    pool.close()
    pool.join()
    return result

# Main execution
if __name__ == "__main__":
    n = int(input("Enter number of lanes: "))
    lanes = []

    for i in range(n):
        cars = int(input("Enter vehicles in lane: "))
        lanes.append(cars)

    # Sequential execution
    start = time.time()
    seq_result = sequential(lanes)
    end = time.time()

    print("\nSequential Results:")
    for r in seq_result:
        print(r)
    print("Sequential Time:", round(end - start, 5), "seconds")

    # Parallel execution
    start = time.time()
    par_result = parallel(lanes)
    end = time.time()

    print("\nParallel Results:")
    for r in par_result:
        print(r)
    print("Parallel Time:", round(end - start, 5), "seconds")
```
